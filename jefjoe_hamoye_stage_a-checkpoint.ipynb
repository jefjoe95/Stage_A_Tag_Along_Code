```python
# Q3 
# Perform a groupby operation on ‘Element’. What is the total number of the sum of Processing in 2017?
fao_data[['Y2017','Element']].groupby(['Element']).sum().loc['Processing']
```

```python
# Q4a
# What is the total number of missing data in 2014
fao_data.isnull().sum()
```

```python
# Q4b
# What is the total number and percentage of missing data in 2014 to 3 decimal places

fao_data.isnull().sum()/fao_data.count()*100
```

```python
# Q5
# What is the total Protein supply quantity in Madagascar in 2015?
fao_data[['Element','Area', 'Y2015']][fao_data['Area'] == 'Madagascar'][fao_data['Element'] == 'Protein supply quantity (g/capita/day)'].sum()
```

```python
# Q6
# Select columns ‘Y2017’ and ‘Area’, Perform a groupby operation on‘Area’. Which of these Areas had the highest sum in 2017?
fao_data[['Area', 'Y2017']].groupby(['Area']).sum().sort_values(by = 'Y2017', ascending=False)
```

```python
# Q8
# Which year had the least correlation with ‘Element Code’?
fao_data.corr()['Element Code'].sort_values()
```
